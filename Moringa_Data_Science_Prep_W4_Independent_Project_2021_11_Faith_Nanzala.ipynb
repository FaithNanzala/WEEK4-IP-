```python
# Procedure 1: Irrelevant Observation
# Data Cleaning Action: Dropping columns 
# Explanation:We do not need them or they do not in fit the context of the problem

Autolib2.drop(['Displayed comment', 'ID', 'Geo point','Address', 'Cars', 'Subscription status', 'year', 'Charge Slots'], axis = 1, inplace = True)
Autolib2
```

```python
# Procedure 2: Outliers
# Data Cleaning Action: checking for outliers using IQR Scores
# Explanation:


Q1 = Autolib2.quantile(0.25)
Q3 = Autolib2.quantile(0.75)

IQR = Q3 - Q1


(Autolib2 < (Q1 - 1.5 * IQR)) |(Autolib2 > (Q3 + 1.5 * IQR))


# The two commented lines of code below can be used to drop the outliers but for now I will not drop them

# Autolib2_no_outlier = Autolib2[~((Autolib2 < (Q1 - 1.5 * IQR)) | (Electric > (Q3 + 1.5 * IQR))).any(axis = 1)]
#Autolib2_no_outlier
```

1.4.5 Uniformity

```python
# Procedure 1: None
# Data Cleaning:None
# Explanation:None

# Hint:
# To compute usage, we will need to understand we have to join successive 
# (in time) measures/counters for a given station, as the difference will tell whether a car was picked up, returned, or nothing happened.

# We are creating a new column where we find the difference in the number of bluecars at the station/counter

Autolib21['Bluecar_Diff'] = Autolib21['Bluecar counter'].diff()
Autolib21.head(10)
```

1.5 Answering Research Questions

```python
#Identify the most popular hour of the day for picking up a shared electric car (Bluecar)
#in the city of Paris over the month of April 2018.

Autolib21_paris = Autolib21[Autolib21['City'] =='Paris'] 

Autolib21_paris.head(10)
```

```python
#What is the most popular hour for returning cars?

Autolib21_paris[Autolib21_paris['Bluecar_Diff'] > 0].groupby('hour')['hour'].count().sort_values(ascending= False)
```

```python
#What station is the most popular?
#Overall?
Autolib21_paris[(Autolib21_paris['Kind'] == 'STATION') & (Autolib21_paris['Status'] == 'ok')].groupby('Public name')['Public name'].count().sort_values(ascending= False)

```

```python
# What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code?

Autolib21_paris[(Autolib21_paris['Kind'] == 'STATION') & (Autolib21_paris['Status'] == 'ok')].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)

# from the Autolib21 dataframe 75015 == Paris/Suffren/2 # therefore NO it doesn't
```

```python
# What postal code is the most popular for picking up Blue cars? Does the most popular station belong to that postal code?

#At the most popular picking hour?

Autolib21_paris[(Autolib21_paris['Kind'] == 'STATION') & (Autolib21_paris['Status'] == 'ok') & (Autolib21_paris['hour'] == 4)].groupby('Postal code')['Postal code'].count().sort_values(ascending= False)

```
